```python
import pandas as pd

fund_master = pd.read_csv("../data/raw/01_fund_master.csv")

print("Shape:", fund_master.shape)
fund_master.head()
```

```python
print("Columns:")
print(fund_master.columns)

print("\nData Types:")
print(fund_master.dtypes)
```

```python
print("Missing Values:")
print(fund_master.isnull().sum())

print("\nDuplicate Rows:")
print(fund_master.duplicated().sum())
```

```python
print("Unique Fund Houses:")
print(fund_master['fund_house'].unique())

print("\nCategories:")
print(fund_master['category'].unique())

print("\nSub Categories:")
print(fund_master['sub_category'].unique())
```

```python
print("DATA QUALITY SUMMARY")
print("-" * 30)
print(f"Rows: {fund_master.shape[0]}")
print(f"Columns: {fund_master.shape[1]}")
print(f"Missing Values: {fund_master.isnull().sum().sum()}")
print(f"Duplicate Rows: {fund_master.duplicated().sum()}")
print(f"Unique Fund Houses: {fund_master['fund_house'].nunique()}")
print(f"Categories: {fund_master['category'].nunique()}")
print(f"Sub Categories: {fund_master['sub_category'].nunique()}")
```
